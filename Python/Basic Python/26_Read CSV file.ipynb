## Read a CSV and Excel File

#### Read a CSV file

```python
import csv

with open('Customers.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

#### Read CSV file having tab delimeter

```python
import csv

with open('Customers.csv', 'r') as file:
    reader = csv.reader(file, delimiter = '\t')
    for row in reader:
        print(row)
```

#### Read CSV files with initial spaces

```python
import csv

with open('Customers.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, skipinitialspace=True)
    for row in reader:
        print(row)
```

#### Read CSV files with quotations

```python
import csv

with open('Customers.csv', 'r') as file:
    reader = csv.reader(file, quoting=csv.QUOTE_ALL, skipinitialspace=True)
    for row in reader:
        print(row)
```
